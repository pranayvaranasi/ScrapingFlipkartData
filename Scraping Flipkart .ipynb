```python
import pandas as pd
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
url = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
urls=[]
for i in range(1,16):
    url=["https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=%i"%(i)]
    urls.append(url[0])
urls


```

Prettifying the page

```python
def prettify(url):
    client = urlopen(url)
    html = client.read()
    client.close()
    pagesoup = soup(html,"html.parser")
    container = pagesoup.find_all("div",{"class":"_75nlfW"})
    return container
con = prettify(urls[0])
print(soup.prettify(con[1]))
```

Creating Dataset

```python
def dataset(containers):
    data={"Name":[],"Sales Price":[],"Actual Price":[],"Rating":[],"Review":[],"Offer":[]}
    for i in range(0,15):
        con = prettify(containers[i])
        for product in con:
            name=product.div.img["alt"]
            data["Name"].append(name)
            try:
                sprice = product.findAll("div",{"class":"Nx9bqj _4b5DiR"})
                sp=sprice[0].text
                data["Sales Price"].append(sp)
            except:
                data["Sales Price"].append("Not launched yet")
            try:
                aprice=product.findAll("div",{"class":"yRaY8j ZYYwLA"})
                ap=aprice[0].text
                data["Actual Price"].append(ap)
            except IndexError:
                print("Same as Sales Price")
                data["Actual Price"].append("Same Price")
            try :
                rating = product.findAll("div",{"class":"XQDdHH"})
                r=rating[0].text
                data["Rating"].append(r)
            except IndexError:
                data["Rating"].append("No Rating")
            try :
                review = product.findAll("span",{"class":"Wphh3N"})
                re = review[0].text
                data["Review"].append(re)
            except IndexError:
                data["Review"].append("No Reviews")
            try :
                offer = product.findAll("div",{"class":"UkUFwK"})
                off = offer[0].text
                data["Offer"].append(off)
            except IndexError:
                data["Offer"].append("No Offer")
        print("done",i)
    return data
dataset = dataset(urls)
dataset


                
                
                
            
```

Converting to Dataframe

```python
df = pd.DataFrame(dataset)
df["Rating"]=pd.to_numeric(df["Rating"])
df.dtypes
```

```python
df[['Total Ratings','Total Reviews']]=df["Review"].str.split("&",1,expand=True)
df["Total Ratings"]=df["Total Ratings"].str.split(" ").str[0]
df["Total Reviews"]=df["Total Reviews"].str.split(" ").str[0]
df.drop("Review",axis=1)



```

Storing in CSV

```python
df.to_csv("Flipkart Mobile Scraped Dataset.csv")
```

Store in MongoDB

```python
import pymongo
def dftomongodb(dataset):
    prettify=pymongo.MongoClient(host='localhost',port=27017)
    database=prettify['FlipkartMobileDataset']
    collection=database['Mobiles']
    for (row,i) in dataset.iterrows():
        Name=i[0]
        Sprice=i[1]
        Aprice=i[2]
        Rating=i[3]
        Offer=i[4]
        TRatings=i[5]
        TReviews=i[6]
        d = {"Name":Name,"Rating":Rating,"Sales_Price":Sprice,"Actual_Price":Aprice,"Offer":Offer,"Total Ratings":TRatings,"Total Reviews":TReviews}
        collection.insert_one(d)
    print("Done")
dftomongodb(df)
        
```
